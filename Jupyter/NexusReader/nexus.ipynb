```python
import h5py
import os

def test_event_time_zero(event_time_zero):
    num = 0
    num2 = 0
    sortable_list = list(event_time_zero)
    sortable_list.sort()
    error_index = []
    for i in range(len(sortable_list) - 1):
        diff = float(sortable_list[i + 1]) - float(sortable_list[i])
        if abs(diff) > 30000000:
            num = num + 1
        if abs(diff) > 50000000:
            error_index.append(i)
            num2 = num2 + 1
    return f"{len(sortable_list)}, {num}, {num2}", error_index

def test_file(path):
    f = h5py.File(path, 'r')
    str, error_index = test_event_time_zero(f["raw_data_1"]["detector_1"]["event_time_zero"])
    if error_index is not []:
        print("   ",f["raw_data_1"]["start_time"][()],f["raw_data_1"]["end_time"][()], error_index)
    print(f"{path}: {str}")
```

```python

path = "../../archive/incoming/hifi/"

directory = os.fsencode(path)

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".nxs"):
        try:
            test_file(path + filename)
        except RuntimeWarning as e:
            print(f"Warning encounters in {filename}: {e}")
        except Exception as e:
            print(f"File {filename} failed {e}")
```
